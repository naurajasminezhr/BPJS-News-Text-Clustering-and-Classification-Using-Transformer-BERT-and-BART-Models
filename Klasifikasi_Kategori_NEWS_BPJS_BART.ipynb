# Klasifikasi Zero-Shot Teks Berita Menggunakan Model BART

Kode ini menggunakan model BART-Large-MNLI dari Hugging Face untuk melakukan klasifikasi teks berita dengan pendekatan Zero-Shot Classification. Zero-shot classification memungkinkan model untuk mengklasifikasikan teks ke dalam kategori tertentu tanpa memerlukan pelatihan ulang pada dataset spesifik.

```python
from transformers import pipeline

# Inisialisasi model untuk klasifikasi teks
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Daftar judul berita
texts = [
    "Tak Mampu Bayar Iuran, Bisakah Menonaktifkan BPJS Kesehatan?",
    "Cara Pindah Faskes BPJS Kesehatan Online",
    "Dirut BPJS: Dokter Asing Boleh Layani Pasien BPJS Kesehatan, asal...",
    "Sejahterakan Pekerja, BPJS Ketenagakerjaan Kucurkan MLT ke Pengembang untuk Bangun 200 Rumah Pekerja"
]

# Daftar kategori berita yang ingin diuji
candidate_labels = ["Kesehatan", "Teknologi", "Ketenagakerjaan", "Hukum", "Ekonomi"]

# Melakukan klasifikasi untuk setiap judul berita
for text in texts:
    result = classifier(text, candidate_labels)
    print(f"Judul: {text}")
    print(f"Kategori: {result['labels'][0]} dengan skor {result['scores'][0]:.2f}")
```

```python
from transformers import pipeline
import pandas as pd
from tqdm import tqdm

# Inisialisasi model untuk klasifikasi teks
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Membaca data dari file CSV
data = pd.read_csv("/content/data_berita_judul - Copy.csv")
texts = data['judul'].tolist()

# Daftar kategori berita yang ingin diuji
candidate_labels = ["Kesehatan", "Teknologi", "Ketenagakerjaan", "Hukum", "Ekonomi"]

# Fungsi untuk klasifikasi satu judul berita
def classify_news_headline(text, labels, classifier):
    result = classifier(text, labels)
    return result['labels'][0], result['scores'][0]

# Membuat list untuk menyimpan hasil klasifikasi
results_list = []

# Melakukan klasifikasi untuk setiap judul berita dengan tqdm untuk progress bar
for text in tqdm(texts, desc="Klasifikasi Sedang Berjalan"):
    category, score = classify_news_headline(text, candidate_labels, classifier)
    results_list.append({"Judul": text, "Kategori": category, "Skor": score})

# Menggabungkan hasil klasifikasi ke dalam DataFrame menggunakan pd.concat()
results = pd.DataFrame(results_list)

# Menyimpan hasil klasifikasi ke file CSV
results.to_csv("hasil_klasifikasi.csv", index=False)

# Menampilkan ringkasan hasil
print(f"Jumlah judul yang diklasifikasikan: {len(results)}")
print(results['Kategori'].value_counts())
```
